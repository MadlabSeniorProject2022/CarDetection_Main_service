```python
from lean_detect import UseModel

logo_detect = UseModel("torch_model/CLD2500.pt", confident=0.6)
```

```python
import cv2
import os
from pathlib import Path
import time
 
# using now() to get current time

def logo_do (coor, img, conf, cls):
    current_time = int(round(time.time() * 1000))
    print(current_time)
    if not os.path.exists("./runs/"):
        os.makedirs("./runs/")
    Path("./runs/ep/").mkdir(parents=True, exist_ok=True)
    crop_img = img[ int(coor[1]) : int(coor[3]), int(coor[0]): int(coor[2])]
    cv2.imwrite("./runs/ep/img" + str(current_time) + ".png", crop_img)
    
```

```python
logo_detect.detect("./field_src/Grey/", do_function=logo_do)
```
