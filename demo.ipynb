```python
from lean_detect import UseModel
```

```python
vehicle_detect = UseModel("yolov7.pt", detect_class=[2, 5, 7])
```

```python
lpd_detect = UseModel("LPD100x.pt", confident=0.5)
```

Example for use 2 stack models.

```python
import cv2

def plate_do (coor, img, conf, cls):
    crop_img = img[ int(coor[1]) : int(coor[3]), int(coor[0]): int(coor[2])]
    cv2.imwrite("./okl.png", crop_img)

def vehicle_do (coor, img, conf, cls):
    crop_img = img[ int(coor[1]) : int(coor[3]), int(coor[0]): int(coor[2])]
    cv2.imwrite("./ok.png", crop_img)
    lpd_detect.detect("./ok.png", do_function=plate_do)
```

```python
import cv2
import os
from pathlib import Path
import time
 
# using now() to get current time

def crop_do (coor, img, conf, cls):
    current_time = int(round(time.time() * 1000))
    print(current_time)
    if not os.path.exists("./runs/"):
        os.makedirs("./runs/")
    Path("./runs/Unseen2/").mkdir(parents=True, exist_ok=True)
    
    crop_img = img[ int(coor[1]) : int(coor[3]), int(coor[0]): int(coor[2])]
    result = cv2.resize(crop_img, (640, 640))
    result = cv2.cvtColor(result , cv2.COLOR_RGB2GRAY)
    cv2.imwrite("./runs/Unseen2/img" + str(current_time) + ".jpg", result)
    
```

```python
vehicle_detect.detect("./clips/use001.MP4" , do_function=crop_do)
```
